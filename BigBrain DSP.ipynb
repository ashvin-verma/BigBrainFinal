# DSP
Loading digital signal processing packages for PSD waterfall and waveform plots

```julia
using DSP, WAV, PyPlot
s, fs = wavread("ZOOM0004.WAV")
```

```julia
plot(0:1/fs:(length(s)-1)/(2*fs), s)
xlabel("Time [s]")
```
